**Introduction**

Histogram is a graphical representation of the intensity distribution of an image. In simple terms, it represents the number of pixels for each intensity value considered.

```
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```
image = cv2.imread("/content/Histogram.jpg")
image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image)
```

```
plt.figure()
plt.title("Grayscale Histogram of uploaded image")
plt.xlabel("grayscale value")
plt.ylabel("# pixels")
hist_previous = plt.hist(image.ravel(),256,[0,256])
plt.show()
```

```
equalized = cv2.equalizeHist(image)
plt.imshow(equalized)
```

```
plt.figure()
plt.title("Grayscale Histogram of equalized image")
plt.xlabel("grayscale value")
plt.ylabel("# pixels")
hist_equalized = plt.hist(equalized.ravel(),256,[0,256])
plt.show()
```

**Summary**

```
images_comparision = np.hstack((image, equalized))
plt.imshow(images_comparision)
```

```
hist_comparision = np.hstack((plt.hist(image.ravel(),256,[0,256]), plt.hist(equalized.ravel(),256,[0,256])))
```

It is possible to equalize the histogram in RGB images by equalize each channel histogram. Histogram Equalization is a computer image processing technique used to improve contrast in images. It accomplishes this by effectively spreading out the most frequent intensity values, i.e. stretching out the intensity range of the image. This method usually increases the global contrast of images when its usable data is represented by close contrast values. This allows for areas of lower local contrast to gain a higher contrast.

**Other possibilities of histogram equalization**


*   Adaptive Histogram Equalization
*   Contrastive Limited Adaptive Equalization



```
lena = cv2.imread("/content/Lena.png")
plt.imshow(lena)
```

```
image_gray = cv2.cvtColor(lena, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray)
```

```
plt.hist(image_gray.ravel(), bins=256, range=(0, 256))
plt.show
```

```
# create a CLAHE object (Arguments are optional).
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
image_after_clahe = clahe.apply(image_gray)
plt.imshow(image_after_clahe)
```

```
plt.hist(image_after_clahe.ravel(), bins=256, range=(0, 256))
plt.show
```
